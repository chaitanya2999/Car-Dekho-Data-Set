```python
final=data.drop(['fuel','seller_type','name','transmission','owner'],axis=1)
```

# Statistical Analysis

```python
features=final.iloc[:,:-1].values
labels=final.iloc[:,-1].values
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
for i in range(1,len(final.columns)-1):
    sel=SelectKBest(chi2,k=i)
    col=sel.fit_transform(features,labels)
    print(col)
```

# Model Building


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
   features,labels,test_size=0.2
)
```

# Training and testing

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
print(classification_report(y_test,y_pred))
```
